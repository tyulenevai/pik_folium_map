```python
import folium


boundary = gpd.read_file('границы_территории_для_поиска_особняка.gpkg')
fin_parcel = gpd.read_file('fin_parcel_v100.geojson')
fin_build = gpd.read_file('fin_build_with_osobnyak.gpkg')


print(fin_parcel.geometry.type.unique())
print(fin_build.geometry.type.unique())
boundary = boundary.to_crs(epsg=4326)
fin_parcel = fin_parcel.to_crs(epsg=4326)
fin_build = fin_build.to_crs(epsg=4326)


m = folium.Map(location=[boundary.geometry.centroid.y.mean(), boundary.geometry.centroid.x.mean()], 
               zoom_start=14, 
               tiles="OpenStreetMap")

boundary_layer = folium.GeoJson(
    boundary,
    name="Границы территории для поиска особняка",
    style_function=lambda x: {
        'color': 'red',        # Цвет края
        'weight': 2,           # Толщина линии
        'fillColor': 'transparent',  # Прозрачная заливка
        'fillOpacity': 0       # Прозрачность заливки
    }
)
boundary_layer.add_to(m)

def create_building_popup(feature):
    descr = feature['Descr']
    area = feature['area']
    name = feature['name_osobnyak']
    link = feature['link']
    return f"""
    <b>Кадастровый номер:</b> {descr}<br>
    <b>Площадь участка, кв. м:</b> {area}<br>
    <b>Особняк:</b> {name}<br>
    <b>Ссылка:</b> <a href="{link}" target="_blank">{link}</a>
    """
buildings_layer = folium.FeatureGroup(name="Кадастровые участки с особняком")
for _, row in fin_build.iterrows():
    popup_text = create_building_popup(row)
    folium.GeoJson(
        data=row['geometry'].__geo_interface__,
        style_function=lambda x: {'color': 'red', 'weight': 2},
        popup=folium.Popup(popup_text, max_width=300)
    ).add_to(buildings_layer)
buildings_layer.add_to(m)

def create_parcel_popup(feature):
    descr = feature['Descr']
    area = round(feature['area'])
    green_type = feature['green_type']
    green_type_desc = "Озелененная территория с деревьями" if green_type == "wool" else "Озелененная территория"
    return f"""
    <b>Кадастровый номер:</b> {descr}<br>
    <b>Площадь участка, кв. м:</b> {area}<br>
    <b>Тип озеленения:</b> {green_type_desc}
    """

parcels_layer = folium.FeatureGroup(name="Кадастровые участки")
for _, row in fin_parcel.iterrows():
    popup_text = create_parcel_popup(row)
    folium.GeoJson(
        data=row['geometry'].__geo_interface__,
        style_function=lambda x: {'color': 'green', 'weight': 2},  # Увеличиваем толщину линий
        popup=folium.Popup(popup_text, max_width=300)
    ).add_to(parcels_layer)
parcels_layer.add_to(m)

folium.LayerControl().add_to(m)
m.save('index.html')
m
```
